```python
import os
import random

# Configuration
base_path = '/content/drive/MyDrive/×ª×•××¨/speech'
subfolders = ['s1', 's2', 'mix_clean']
scp_output_dir = '/content/SepReformer/data/custom'
os.makedirs(scp_output_dir, exist_ok=True)

# 1. Get list of file names (assuming they are identical across s1, s2, and mix)
# We sort them first to ensure indices match across all subfolders
files = sorted([f for f in os.listdir(os.path.join(base_path, 's1')) if f.endswith('.wav')])

# 2. Shuffle and Split (80/20)
random.seed(42) # For reproducibility
random.shuffle(files)
split_idx = int(len(files) * 0.8)

train_files = files[:split_idx]
val_files = files[split_idx:]

# 3. Generate the SCP files
def create_scp(file_list, set_name):
    for folder in subfolders:
        output_path = os.path.join(scp_output_dir, f"{set_name}_{folder}.scp")
        with open(output_path, 'w') as f:
            for filename in file_list:
                full_path = os.path.join(base_path, folder, filename)
                # ID format: filename (must be unique)
                f.write(f"{filename} {full_path}\n")
        print(f"Created {output_path} with {len(file_list)} files.")

create_scp(train_files, "train")
create_scp(val_files, "val")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class AttentionSpeakerSplit(nn.Module):
    def __init__(self, dim, num_speakers=2, num_heads=8, dropout=0.05):
        """
        Args:
            dim (int): Feature dimension (e.g., 256)
            num_speakers (int): Number of speakers to split into (e.g., 2)
            num_heads (int): Number of attention heads
            dropout (float): Dropout probability
        """
        super(AttentionSpeakerSplit, self).__init__()

        self.dim = dim
        self.num_speakers = num_speakers

        # Learnable Speaker Queries (Q)
        # Shape: (Num_Speakers, Dim)
        self.speaker_queries = nn.Parameter(torch.randn(num_speakers, dim))

        # Cross-Attention Module
        # We use batch_first=True for easier handling of (Batch, Seq, Dim) tensors
        self.mha = nn.MultiheadAttention(embed_dim=dim,
                                         num_heads=num_heads,
                                         dropout=dropout,
                                         batch_first=True)

        # Normalization and Dropout for stability
        self.norm = nn.LayerNorm(dim)
        self.dropout = nn.Dropout(dropout)

        # Initialize parameters
        self._init_weights()

    def _init_weights(self):
        nn.init.xavier_uniform_(self.speaker_queries)

    def forward(self, x):
        """
        Args:
            x: Input tensor of shape (Batch, Dim, Time)
        Returns:
            out: Output tensor of shape (Batch, Num_Speakers, Dim, Time)
        """
        B, D, T = x.shape

        # 1. Prepare Keys and Values from Input
        # Input x is (Batch, Dim, Time). MHA expects (Batch, Time, Dim).
        keys_values = x.permute(0, 2, 1)

        # 2. Prepare Queries from Learnable Parameters
        # Queries are (Num_Speakers, Dim). Expand to (Batch, Num_Speakers, Dim).
        queries = self.speaker_queries.unsqueeze(0).expand(B, -1, -1)

        # 3. Apply Cross-Attention
        # Query: (Batch, Num_Speakers, Dim) -> Looking for speaker info
        # Key/Val: (Batch, Time, Dim) -> The audio context
        # Output: (Batch, Num_Speakers, Dim) -> Global Speaker Vectors
        attn_out, _ = self.mha(query=queries, key=keys_values, value=keys_values)

        # Apply Residual block processing (Dropout + Norm)
        attn_out = self.norm(self.dropout(attn_out))

        # 4. Broadcast and Combine (Residual Connection)
        # We combine the Global Speaker Vectors with the original Local Time Features.

        # attn_out: (Batch, Num_Speakers, Dim) -> (Batch, Num_Speakers, Dim, 1)
        global_speaker_info = attn_out.unsqueeze(-1)

        # x: (Batch, Dim, Time) -> (Batch, 1, Dim, Time)
        local_time_features = x.unsqueeze(1)

        # Result: (Batch, Num_Speakers, Dim, Time)
        # This adds the speaker-specific bias to every time step of the original input.
        out = local_time_features + global_speaker_info

        return out
```
